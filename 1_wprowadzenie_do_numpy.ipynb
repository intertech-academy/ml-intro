## Przykład 1 - Wykres funkcji jednej zmiennej
Na początek spróbujemy narysować wykres funkcji jednej zmiennej $y=f(x)$. Będziemy do tego potrzebować dwóch tablic: jednej z wartościami zmiennej $x$ i drugą z wartościami $y$.

Na początek spróbujmy wygenerować i wyświetlić dwie takie tablice:

```
x = np.array([1, 2, 3])
y = x**2

print("x: "+str(x))
print("y: "+str(y))
```

Za pomocą funkcji np.array() możemy zwykłą listę Pythona zamienić na obiekt klasy ndarray z modułu NumPy. NumPy znacznie ułatwia obliczenia na tych obiektach. Zwróć uwagę np. na to, że możemy podnieść do kwadratu wszystkie elementy listy za pomocą jednej instrukcji.

Tworzenie tablic wypisują ręcznie ich elementy nie jest zbyt wygodne. Użyjmy funkcji linspace, żeby wygenerować tablicę z dwudziestoma równoodległymi liczbami z zakresu od $0$ do $2\pi$: 



```
x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)

print("x: "+str(x))
print("y: "+str(y))
```

Przeglądanie tak dużych tablic poprzez wyświetlanie ich w konsoli też nie jest zbyt wygodne. Użyjmy teraz modułu PyPlot do wygenerowania wykresu funkcji $f(x)=\sin(x)$:

```
plt.plot(x, y)
```

## Przykład 2 - Wykres funkcji dwóch zmiennych

Za pomocą NumPy możemy też tworzyć tablice wielowymiarowe. Spróbujmy użyć tej możliwości do wygenerowania wykresu funkcji $z=\cos(x)\sin(y)$. Na początek potrzebujemy dwuwymiarowe "mapy" wartości $x$ i $y$. Przyda się do tego funkcja meshgrid:

```
xx, yy = np.meshgrid(np.array([1,2,3]), np.array([-1,-2,-3]))

print("xx: "+str(xx))
print("yy: "+str(yy))
```

Skoro umiemy tworzyć tablice dwuwymiarowe to możemy teraz obliczyć wartości funkcji. Do narysowania jej wykresu (w formie kolorowej mapy) użyjemy funkcji pcolor:

```
xx, yy = np.meshgrid(np.linspace(0, 2*np.pi, 100), np.linspace(0, 2*np.pi, 100))
zz = np.cos(xx)*np.sin(yy)

plt.pcolor(xx, yy, zz)
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
```

## Przykład 3 - Zbiór Mandelbrota

Teraz możemy zrobić coś bardziej widowiskowego. Spróbujmy narysować zbiór Mandelbrota. Zrozumienie czym jest ten zbiór wymaga znajomości liczb zespolonych. Jeżeli chcesz to możesz pominąć poniższe wyjaśnienie i przejrzeć jak działa kod generujący ten zbiór.

### Czym jest zbiór Mandelbrota
Wyobraź sobie, że tworzymy ciąg liczb zespolonych ustalając jego pierwszy element na zero:
$$z_0 = 0,$$
a następnie każdy kolejny element obliczamy podnosząc poprzedni do kwadratu i dodając pewną stałą liczbę zespoloną $p$:
$$z_{n+1}=z_n^2+p.$$

Dla niektórych wartości $p$ moduł $z_n$ będzie rósł do nieskończoności, a dla niektórych nie. Zbiór Mandelbrota jest zbiorem tych wartości $p$, dla których nasz ciąg nie jest rozbieżny.

```
#Generujemy zbiór parametrów p
re, im = np.meshgrid(np.linspace(-2, 1, 512), np.linspace(-1, 1, 512))
p = re + 1j*im

#Inicjalizujemy wartości ciągu
z = np.zeros(p.shape)
res = np.zeros(p.shape)

#Przeprowadzamy iteracje
no_iter = 200
z_max = 5
for it in range(no_iter):
  z = z**2 + p

  #Zaznaczamy na obrazie wynikowym punkty, dla których ciąg jest rozbieżny
  ind = np.argwhere(np.logical_and(np.abs(z)>z_max, res==0))
  res[ind[:, 0], ind[:, 1]] = it+1

#Rysujemy wynik
plt.pcolor(re, im, res)
plt.gca().set_aspect('equal')
```

## Przykład 4 - zbiory Julii

Zbiory Julii są bardzo blisko powiązane ze zbiorem Mandlebrota tym razem wybieramy pewną liczbę zespoloną $c$ i przeprowadzamy iterację:
$$z_0=p,$$
$$z_{n+1}=z_n^2+c.$$

Do zbioru Julii odpowiadającemu parametrowi $c$ należą te liczby zespolone $p$, dla których ciąg nie jest rozbieżny.

```
#Generujemy zbiór parametrów p
re, im = np.meshgrid(np.linspace(-1.5, 1.5, 512), np.linspace(-1.5, 1.5, 512))
p = re + 1j*im

#Inicjalizujemy wartości ciągu
z = p
res = -10*np.ones(p.shape)

#Wybieramy parametr c
c = -0.73+0.19j

#Przeprowadzamy iteracje
no_iter = 250
z_max = 10
for it in range(no_iter):
  z = z**2 + c
  ind = np.argwhere(np.logical_and(np.abs(z)>z_max, res<0))
  res[ind[:, 0], ind[:, 1]] = it

#Rysujemy wynik
plt.pcolor(re, im, res)
plt.gca().set_aspect('equal')
```
