# **Wprowadzenie do NumPy i Matplotlib**

```
import numpy as np
import matplotlib.pyplot as plt
```

## Przykład 1 - Wykres funkcji jednej zmiennej

```
x = np.array([1, 2, 3])
y = x**2

print("x: "+str(x))
print("y: "+str(y))
```

```
x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)

print("x: "+str(x))
print("y: "+str(y))
```

```
plt.plot(x, y)
```

## Przykład 2 - Wykres funkcji dwóch zmiennych

```
xx, yy = np.meshgrid(np.array([1,2,3]), np.array([-1,-2,-3]))

print("xx: "+str(xx))
print("yy: "+str(yy))
```

```
xx, yy = np.meshgrid(np.linspace(0, 2*np.pi, 100), np.linspace(0, 2*np.pi, 100))
zz = np.cos(xx)*np.sin(yy)

plt.pcolor(xx, yy, zz)
plt.xlabel('x')
plt.ylabel('y')
```

## Przykład 3 - Zbiór Mandelbrota

```
re, im = np.meshgrid(np.linspace(-2, 1, 512), np.linspace(-1, 1, 512))
p = re + 1j*im
```

```
z = np.zeros(p.shape)
res = np.zeros(p.shape)

no_iter = 200
z_max = 5
for it in range(no_iter):
  z = z**2 + p
  ind = np.argwhere(np.logical_and(np.abs(z)>z_max, res==0))
  res[ind[:, 0], ind[:, 1]] = it+1
  
plt.pcolor(re, im, res)
plt.gca().set_aspect('equal')
```

## Przykład 4 - zbiory Julii

```
re, im = np.meshgrid(np.linspace(-1.5, 1.5, 512), np.linspace(-1.5, 1.5, 512))
p = re + 1j*im
```

```
z = p

c = -0.73+0.19j
res = -10*np.ones(p.shape)

no_iter = 500
z_max = 10
for it in range(no_iter):
  z = z**2 + c
  ind = np.argwhere(np.logical_and(np.abs(z)>z_max, res<0))
  res[ind[:, 0], ind[:, 1]] = it

plt.pcolor(re, im, res)
plt.gca().set_aspect('equal')
```
